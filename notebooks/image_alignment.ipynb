```python
import cv2
import os
import  numpy as np
dmr_dir = "/home/group-cvg/cvg-students/sshah/d_m_r"
```

```python
im1 = cv2.imread(os.path.join(dmr_dir,'data/MOT17/train/MOT17-02-FRCNN/img1/000010.jpg'))
im2 = cv2.imread(os.path.join(dmr_dir,'data/MOT17/train/MOT17-04-FRCNN/img1/000012.jpg'))
```

```python
im1_gray = cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
im2_gray = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
```

```python
sz = im1.shape
```

```python
warp_mode = cv2.MOTION_AFFINE
```

```python
warp_matrix = np.eye(2, 3, dtype=np.float32)
```

```python
number_of_iterations = 200
termination_eps = 1e-10
```

```python
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, number_of_iterations,  termination_eps)
```

```python
(cc, warp_matrix) = cv2.findTransformECC(im1_gray,im2_gray,warp_matrix, warp_mode, criteria)
```

```python
im2_aligned = cv2.warpAffine(im2_gray, warp_matrix, (sz[1],sz[0]), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
cv2.imshow("Image 1", im1_gray)
cv2.imshow("Image 2", im2_gray)
cv2.imshow("Aligned Image 2", im2_aligned)
cv2.waitKey(0)
```
